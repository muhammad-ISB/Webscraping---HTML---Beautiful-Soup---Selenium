```python
soup = BeautifulSoup(html_data, 'html.parser')
events = soup.find_all('div', class_='events-details')  # Find all event divs

data = []

for entry in events:
    company_name = entry.find('h3').text.strip() if entry.find('h3') else 'Unknown'

    location_elements = entry.find('li', class_='location').find_all('span')
    location = ", ".join([element.text.strip() for element in location_elements])

    ideation_elem = entry.find('li', text=lambda text: text and text.startswith('Ideation'))
    ideation_text = ideation_elem.text.strip() if ideation_elem else 'Not specified'

    it_services_elem = entry.find('span', class_='dept')
    it_services_text = it_services_elem.text.strip() if it_services_elem else 'Not specified'

    data.append({
        'Company Name': company_name,
        'Location': location,
        'Ideation': ideation_text,
        'IT Services': it_services_text
    })

# Create a DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
df
```

```python
# Export to CSV
df.to_csv('events_data.csv', index=False)
```

```python
data = []  # List to store each event's data

for event in events:
    # Extract the company name
    h3_text = event.find('h3').text.strip() if event.find('h3') else 'Unknown'

    # Extract location details
    location_elements = event.find('li', class_='location').find_all('span')
    if len(location_elements) >= 2:
        location = location_elements[0].text.strip()  # City
        state = location_elements[1].text.strip()  # State
    else:
        location = 'Unknown'
        state = 'Unknown'

    # Extract department, default to 'Unknown' if not found
    dept_element = event.find('span', class_='dept')
    department = dept_element.text.strip() if dept_element else 'Unknown'

    # Extract highlighted text (like Ideation, Validation, etc.)
    highlighted_texts = [li.text.strip() for li in event.find_all('span', class_='highlighted-text')]
    highlighted_text = highlighted_texts[0] if highlighted_texts else 'Unknown'

    # Append data for this event to the list
    data.append({
        'Company Name': h3_text,
        'City': location,
        'State': state,
        'Department': department,
        'Highlighted Text': highlighted_text
    })

# Create a DataFrame
df = pd.DataFrame(data)
df
```
